A **função describe()** do pandas em Python é usada para gerar estatísticas descritivas de um DataFrame ou Series. Essas estatísticas ajudam a entender a distribuição dos dados e são diretamente relacionadas ao boxplot, que é uma forma gráfica de visualizar essas estatísticas.
(ou seja, ele te dá um resumo numérico das colunas com valores quantitativos).


| Linha | O que significa | Importância prática |
| ----- | --------------- | ------------------- |
| **count** | Quantidade de valores **não nulos** (dados válidos) em cada coluna | Mostra se há dados faltando (valores ausentes) |
| **mean** | Média aritmética | Mostra a **tendência central** dos dados |
| **std** | Desvio padrão | Mede **quanto os valores variam** em relação à média |
| **min** | Valor mínimo | Mostra o menor valor registrado |
| **25% (1º quartil)** | Q1 - Valor abaixo do qual estão 25% dos dados | Útil para detectar **distribuição** dos valores |
| **50% (mediana)** | Q2 - Valor do meio da distribuição | Mostra o ponto central real (menos sensível a outliers que a média) |
| **75% (3º quartil)** | Q3 - Valor abaixo do qual estão 75% dos dados | Ajuda a entender **dispersão** |
| **max** | Valor máximo | Mostra o maior valor registrado |

Ele traz um resumo rápido e estatístico dos dados numéricos, e serve para:

* Detectar valores ausentes (se o count varia)
* Identificar outliers (valores muito distantes da média)
* Avaliar a distribuição (comparando média, mediana e quartis)
* Entender o intervalo de variação (min e max)
* Ajudar a definir estratégias de limpeza e normalização de dados

ps. Caso haja colunas categoricas, podemos usar:
`.describe(include='all')`

Assim ele mostra estatísticas também para colunas de texto:

* unique -> quantos valores únicos
* top -> valor mais frequente
* freq -> frequência desse valor mais comum

## **boxplot** (ou diagrama de caixa)

É a forma visual de resumir a distribuição de um conjunto de dados numéricos. Ele representa visualmente as cinco estatísticas do *describe()*:


      |---------|=======|=======|---------|        ○
      min      Q1       Q2      Q3        max     Outlier
                \_______|_______/
                     caixa

**No boxplot:**

* A caixa vai do Q1 (25%) até o Q3 (75%), ou seja, o tamanho da caixa é dado pelo IQR (Intervalo Interquartil): IQR = Q3 - Q1
* A linha dentro da caixa é a mediana (Q2 ou 50%)
* As linhas (bigodes / whiskers) se estendem até o min e max, ou até limites chamados de limites de detecção de outliers (inferior -> Q1 - 1.5 x IQR e superior -> Q3 + 1.5 x IQR)
* *Outliers* são pontos fora desses limites e são plotados individualmente

Ou seja, O boxplot mostra de forma clara:
* a tendência central (onde estão a maioria dos dados),
* a dispersão (o quanto os valores variam),
* e os outliers (valores fora do padrão).

```python
# Visualising 'CarComapny' variable aginst dependent variable 'price' via boxplot

plt.figure(figsize=(18,12))

order = sorted(CarData['CarCompany'].unique())
palette = list(plt.cm.hsv(np.linspace(0, 1, len(order))))

sns.boxplot(
    x='CarCompany',
    y='price',
    hue='CarCompany',
    data=CarData,
    order=order,
    hue_order=order,
    palette=palette,
    dodge=False
)

plt.xticks(rotation=45)
plt.legend([],[], frameon=False)
plt.show()
```

## Como interpretar o gráfico de boxplot

*- O que cada elemento significa*:

Para cada marca de carro no eixo X (como bmw, honda, toyota, etc.), nós temos uma “caixa” vertical representando a distribuição dos preços:

* Quanto maior a caixa, maior a variação nos preços dessa marca.
* “Bigodes” (linhas que saem da caixa)
→ Mostram a faixa de preços mais comuns (sem considerar os outliers).
* Pontos fora da caixa (outliers)
→ Representam preços que fogem muito do padrão da marca — podem ser carros muito mais caros (luxo) ou muito mais baratos (básicos).

*- Como interpretar o gráfico:*

* BMW e Jaguar, por exemplo, têm caixas altas, medianas elevadas e variação ampla → indicam que essas marcas têm preços geralmente altos e variados.

* Chevrolet, Dodge e Honda têm caixas pequenas e localizadas em valores baixos → carros dessas marcas tendem a ter preços mais acessíveis e consistentes.

* Audi e Volvo têm medianas bem mais altas que a média das outras → são marcas de carros premium.

* Os outliers (bolinhas) podem representar modelos esportivos ou de luxo fora da curva — por exemplo, um Porsche muito mais caro que o resto da categoria.

```python
# Visualising Categorical variables through boxplots

fig, axes = plt.subplots(1, 4, figsize=(20,5))

sns.boxplot(x='fueltype', y='price', hue='fueltype', data=CarData, ax=axes[0])
sns.boxplot(x='cylindernumber', y='price', hue='cylindernumber', data=CarData, ax=axes[1])
sns.boxplot(x='enginelocation', y='price', hue='enginelocation', data=CarData, ax=axes[2])
sns.boxplot(x='doornumber', y='price', hue='doornumber', data=CarData, ax=axes[3])

plt.xticks(rotation=45)
plt.legend([],[], frameon=False)

plt.tight_layout()
plt.show()
```

```python
#creating a dict 'company price' where key = car's company and value = median of their price

company_price = CarData.groupby('CarCompany')['price'].median().to_dict()
```

```python
# division in the buckets of low i.e. below 10000, medium i.e. range b/w 10000 and 20000 and high i.e. above 20000

company_price_bucketed = {}

for company, price in company_price.items():
    if price < 10000:
        company_price_bucketed[company] = 'low'
    elif 10000 <= price <= 20000:
        company_price_bucketed[company] = 'medium'
    else:
        company_price_bucketed[company] = 'high'

CarData['CarCompanyCategory'] = CarData['CarCompany'].map(company_price_bucketed)

CarData.CarCompanyCategory.unique()
```

```python
# creating dummy variable for categorical variables
CarData = CarData.drop(columns=['CarName'])
categorical_cols=['fueltype', 'aspiration', 'doornumber', 'carbody',
                  'drivewheel', 'enginelocation', 'enginetype','cylindernumber',
                  'fuelsystem', 'CarCompany', 'CarCompanyCategory']

CarData = pd.get_dummies(CarData, columns=categorical_cols, dtype=int)

#checking the dataset
CarData.head()
```

## **random_state**

O random_state é um parâmetro opcional que você passa em funções que envolvem aleatoriedade (como o `train_test_split`, mas também em outras, como `shuffle`, `KFold`, ou modelos de aprendizado que usam inicialização aleatória).

Ele funciona como uma “semente” (seed) que controla o comportamento do gerador de números aleatórios do Python.

*- Como funciona **SEM** random_state:*

Sem o random_state, o *train_test_split* escolhe aleatoriamente quais amostras vão para treino e quais vão para teste — e cada vez que você roda, essa divisão pode ser diferente.

Isso significa que o modelo pode:

* Ter resultados um pouco diferentes a cada execução (porque os dados de treino/teste mudam);
* Ficar difícil de reproduzir o mesmo resultado depois.

*- Como funciona **COM** random_state:*

Ele funciona como uma “semente” (seed) que controla o comportamento do gerador de números aleatórios do Python.

Ou seja:

* A divisão será sempre a mesma, toda vez que rodar o código.
* Se alguém pegar esse código e usar o mesmo número no random_state (por exemplo: 42), terá exatamente o mesmo resultado.

*- Dicas:*

* Usar o mesmo random_state (como 42, 0, 123, etc.) é uma boa prática em notebooks, projetos de aprendizado ou relatórios, pois garante reprodutibilidade.
* Em produção ou experimentos automatizados, podemos não definir o random_state para permitir variação e testar a robustez do modelo.

```python
# importing required library
from sklearn.model_selection import train_test_split

# We specify this so that the train and test data set always have the same rows, respectively
CarData_train, CarData_test = train_test_split(CarData, test_size=0.2, random_state=42)

# shape of the train and test sets
print(CarData_train.shape)
print(CarData_test.shape)
```

```python
#importing required function for scaling
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
```

```python
target = 'price'

X_train = CarData_train.drop(columns=[target])
y_train = CarData_train[target]

X_test = CarData_test.drop(columns=[target])
y_test = CarData_test[target]

X_train_scaler = scaler.fit_transform(X_train)
X_test_scaler = scaler.transform(X_test)
```

```python
# Train your model
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X_train_scaler, y_train)
```

```python
# Making predictions
predicted_price = model.predict(X_test_scaler)
predicted_price = np.maximum(predicted_price, 0)
```

```python
# Show predictions
predicted_price
```

```python
# Plotting y_test and y_pred to understand the spread.
plt.figure(figsize=(8,6))
plt.scatter(y_test, predicted_price, alpha=0.7)
plt.xlabel("y_test")
plt.ylabel("y_pred")
plt.title("y_test vs y_pred")
plt.show()
```

```python
from sklearn.metrics import r2_score

print(r2_score(y_test, predicted_price))
```
